```python
data.Stay_In_Current_City_Years=data.Stay_In_Current_City_Years.replace("4+",'4').astype(int)
data.Stay_In_Current_City_Years.unique()
```

```python
import matplotlib.pyplot as plt

df = data

fig, axs = plt.subplots(2, 3, figsize=(18, 10))  

# Subplot for Gender
axs[0, 0].pie(df.groupby("Gender")["Purchase"].sum(), labels=df.groupby("Gender")["Purchase"].sum().index, autopct="%1.1f%%")
axs[0, 0].set_title("Purchases by Gender")

# Subplot for Age
axs[0, 1].pie(df.groupby("Age")["Purchase"].sum(), labels=df.groupby("Age")["Purchase"].sum().index, autopct="%1.1f%%")
axs[0, 1].set_title("Purchases by Age")

# Subplot for Occupation
axs[0, 2].pie(df.groupby("Occupation")["Purchase"].sum(), labels=df.groupby("Occupation")["Purchase"].sum().index, autopct="%1.1f%%")
axs[0, 2].set_title("Purchases by Occupation")

# Subplot for City Category
axs[1, 0].pie(df.groupby("City_Category")["Purchase"].sum(), labels=df.groupby("City_Category")["Purchase"].sum().index, autopct="%1.1f%%")
axs[1, 0].set_title("Purchases by City Category")

# Subplot for Stay in Current City Years
axs[1, 1].pie(df.groupby("Stay_In_Current_City_Years")["Purchase"].sum(), labels=df.groupby("Stay_In_Current_City_Years")["Purchase"].sum().index, autopct="%1.1f%%")
axs[1, 1].set_title("Purchases by Stay in Current City Years")

# Subplot for Marital Status
axs[1, 2].pie(df.groupby("Marital_Status")["Purchase"].sum(), labels=df.groupby("Marital_Status")["Purchase"].sum().index, autopct="%1.1f%%")
axs[1, 2].set_title("Purchases by Marital Status")

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Fill NaN values in 'Marital_Status' with a default value 
data['Marital_Status'] = data['Marital_Status'].fillna(0)  


# Verify columns and data
print(data.columns)
print(data[['Gender', 'Marital_Status', 'Purchase']].head())

# Group by Gender and Marital Status
M_UM_P = data.groupby(['Gender', 'Marital_Status']).agg({'Purchase': sum}).reset_index()

# Verify the grouped DataFrame
print(M_UM_P)

# Data for top 10 products purchased by unmarried and married customers
UM = data[data['Marital_Status'] == 0].value_counts('Product_ID').nlargest(10)
M = data[data['Marital_Status'] == 1].value_counts('Product_ID').nlargest(10)

plt.figure(figsize=(18, 10))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

# Subplot 1: Purchases by Marital Status and Gender
plt.subplot(2, 2, 1)
sns.barplot(x='Marital_Status', y='Purchase', hue='Gender', data=M_UM_P)
plt.title("Purchases by Gender and Marital Status")

# Subplot 2: Top 10 Products for Unmarried Customers
plt.subplot(2, 2, 3)
UM.plot(kind='bar', title='Top 10 Products for Unmarried Customers', xlabel='Product ID', ylabel='Count')

# Subplot 3: Top 10 Products for Married Customers
plt.subplot(2, 2, 4)
M.plot(kind='bar', title='Top 10 Products for Married Customers', xlabel='Product ID', ylabel='Count')

plt.show()
```

```python

# Which product customer is buying the most
p_ids = data.value_counts('Product_ID').sort_values(ascending=False).head(50)

plt.figure(figsize=(12, 8))
sns.barplot(x=p_ids.values, y=p_ids.index)
plt.xlabel('Number of Purchases')
plt.ylabel('Product ID')
plt.title('Top 50 Most Purchased Products')
plt.show()
```
